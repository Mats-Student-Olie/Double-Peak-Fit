```python
S2base = r'C:\Users\mbgnwob2\Dropbox (The University of Manchester)\2. Project\Python Script\Single Peak Fitting Script ORIGINAL - Copy\Test_data_Sample2_tension_strain'
S2identifyer = ['0', '5', '10', '20', '30', '42', '54', '57']
S2identifyer = np.char.zfill(S2identifyer, 5)

S2spectrums = []
S2spectrums.append(np.loadtxt(str(S2base + '\\' + prefix + S2identifyer[0] + ftype), usecols = 0))

for i in S2identifyer:
    fname = str(S2base + '\\' + prefix + i + ftype)
    data = np.loadtxt(fname, usecols = 1)
    S2spectrums.append(data)

fig, ax  = plt.subplots(figsize=(18, 10))
plt.plot(S2spectrums[0], S2spectrums[1]/7, label = 'S2')
plt.plot(spectrums[0], spectrums[1], label = 'S3')
plt.legend()
ax.set_yscale('log')
ax.set_xlim(2.5, 11)
ax.set_ylim(100)
ax.axvline(3.15217, ls = '--', color = 'b') #(0002)
ax.axvline(2.94063, ls = '--', color = 'r') #(111)
ax.axvline(4.80291, ls = '--', color = 'r') #(220)
ax.axvline(5.63253, ls = '--', color = 'r') #(311)
ax.set_ylabel('Counts')
ax.set_xlabel('two theta')
ax.set_title('ANSTO Sample2 vs Sample3')
#fig.savefig(out_dir + 'S2vsS3')
```
